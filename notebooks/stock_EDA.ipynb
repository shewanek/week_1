## 4. Calculating Financial Metrics

This section describes the steps to calculate financial metrics for stock data using Python and Pandas. The metrics include Daily Returns, Rolling Volatility, Moving Average, and Cumulative Returns. These metrics are calculated for each stock symbol individually.

### Steps for Calculating Financial Metrics

1. **Sort Data by Date:**
   - Ensure that the data is sorted by `Date` and `stock_symbol` for consistency.

2. **Iterate Over Each Stock Symbol:**
   - For each unique `stock_symbol`, filter the data to create a subset specific to that stock.

3. **Calculate Daily Returns:**
   - Use the Pandas `pct_change()` function to calculate the daily returns.
   - `df['Daily Returns'] = df['Close'].pct_change()`

4. **Calculate Rolling Volatility:**
   - Calculate the rolling standard deviation of daily returns over a 20-day window.
   - `df['Rolling Volatility'] = df['Daily Returns'].rolling(window=20).std()`

5. **Calculate Moving Average:**
   - Calculate the 20-day moving average of the closing prices.
   - `df['Moving Average'] = df['Close'].rolling(window=20).mean()`

6. **Calculate Cumulative Returns:**
   - Calculate the cumulative returns by taking the cumulative product of daily returns and subtracting 1.
   - `df['Cumulative Returns'] = (1 + df['Daily Returns']).cumprod() - 1`

7. **Combine Results:**
   - Concatenate the data for each stock symbol into a single DataFrame.

8. **Return the Resulting DataFrame:**
   - The final DataFrame contains all the calculated financial metrics for each stock symbol.
